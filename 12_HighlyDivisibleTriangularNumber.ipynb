# Highly divisible triangular number
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 
 3: 1,3
 
 6: 1,2,3,6
 
10: 1,2,5,10

15: 1,3,5,15

21: 1,3,7,21

28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
def NumberOfDivisors(number):
    nod = 0
    sqrt = round(math.sqrt(number))
    for i in range(1, sqrt):
        if(number % i == 0):
            nod += 2
            
    # Correction if the number is a perfect square
    if (sqrt * sqrt == number):
        nod -= 1
 
    return nod
```

```python
def getTriangleNumber3(nrDivizors):
    number = 0
    i = 1
    while(NumberOfDivisors(number) < nrDivizors):
        number += i
        i += 1
    return number
```

```python
print("first triangle number with 5 divizors is: {0}".format(getTriangleNumber3(5)))
```

```python
print("first triangle number with 100 divizors is: {0}".format(getTriangleNumber3(100)))
```

```python
print("first triangle number with 500 divizors is: {0}".format(getTriangleNumber3(500)))
```
